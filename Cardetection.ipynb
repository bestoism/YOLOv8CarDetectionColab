```python
# Load model YOLOv8
model = YOLO("yolov8n.pt")
```

```python
from google.colab import files

uploaded = files.upload()

video_path = list(uploaded.keys())[0]
print("Video berhasil diupload:", video_path)
```

```python
# Video yang sudah diupload sebelumnya
video_path = "car-video.mp4"
cap = cv2.VideoCapture(video_path)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Output video
out = cv2.VideoWriter("hasil_deteksi.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Proses frame demi frame
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Deteksi objek
    results = model(frame)

    # Tambahkan anotasi
    annotated_frame = results[0].plot()

    # Tulis frame ke video hasil
    out.write(annotated_frame)

# Tutup semua proses
cap.release()
out.release()
print("Video selesai diproses dan disimpan sebagai hasil_deteksi.mp4")

files.download("hasil_deteksi.mp4")
```
